```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Loading the dataset
file_path = '/content/ratings_Beauty[1].csv'
df = pd.read_csv(file_path)
```

```python
print("Dataset Overview:")
print(df.info())
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python
# Checking for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Checking for duplicates
print("\nDuplicate rows:")
print(df.duplicated().sum())

```

```python
# Analyzing ratings distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Rating', palette='viridis')
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Analyzing unique users and products
unique_users = df['UserId'].nunique()
unique_products = df['ProductId'].nunique()
print(f"\nNumber of unique users: {unique_users}")
print(f"Number of unique products: {unique_products}")

```

```python
# Filtering data for active users
user_ratings_count = df.groupby('UserId').size()
active_users = user_ratings_count[user_ratings_count > 10].index
filtered_df = df[df['UserId'].isin(active_users)]
print(f"\nFiltered dataset contains {filtered_df.shape[0]} rows.")

# Saving filtered data for recommendation system
filtered_df.to_csv('/content/ratings_Beauty[1].csv', index=False)
print("Filtered data saved for recommendation system.")

```
